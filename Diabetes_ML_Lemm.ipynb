Load the data and split the data

I chose to analyze the BMI metric first because from what I have read, weight is the biggest deciding factor as to whether a person will develop diabetes or not

```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd

# Loading the diabetes dataset
db_dataset = load_diabetes()

# Extracting the features and the target
X = db_dataset.data
y = db_dataset.target

# Choosing a feature to predict the target
# Selecting the BMI feature (third column in the dataset)
X_feature = X[:, 2:3]

# Splitting the data into training and test sets
# Using the last 20 rows for the test set
X_train, X_test, y_train, y_test = train_test_split(X_feature, y, test_size=20, shuffle=False)

# Output the shapes of the datasets to verify the split
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

Instantiating the Linear Regression Model 

```python
from sklearn.linear_model import LinearRegression

# Instantiating the linear regression model
lin_reg = LinearRegression()

# Training the model with the single feature (BMI) and target sets
lin_reg.fit(X_train, y_train)
```

Predictions 

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Making predictions on the training set
predictions = lin_reg.predict(X_train)

# Printing the first 10 predictions
first_10_predictions = predictions[:10]

# Calculating the root mean squared error (RMSE) of the model on the training set
rmse = np.sqrt(mean_squared_error(y_train, predictions))

# Getting the feature coefficient (BMI)
feature_coefficient = lin_reg.coef_[0]

first_10_predictions, feature_coefficient, rmse

```

The coeffecient of 938.237 is a positive coeffecient indicating that there is a positive relationship between BMI and Disease Progression - this also indicates for predictions that a 1 unit increase in BMI leads to 938 unit increase in the Disease Progression Number

```python
import matplotlib.pyplot as plt

# Scatter plot of the feature (BMI) vs. disease progression
plt.scatter(X_train, y_train, color='blue', label='Training data')

# Plotting the regression line
plt.plot(X_train, lin_reg.predict(X_train), color='red', label='Regression Line')

# Adding labels and title
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.title('Linear Regression on Diabetes Dataset')

# Adding a legend
plt.legend()

# Showing the plot
plt.show()

```
